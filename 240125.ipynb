```python
realres=[]
res=[]
cnt=1
temp=''
def makekmer(a) :
    global res, cnt, temp
    nt=["A","T","G","C"]
    
    if cnt==1 :
        for x in range(4) :
            res.append(nt[x])
    else :
        for xx in range(len(res)) :
            for y in range(4) :
                temp=res[xx]+nt[y]
                
                if temp not in res :
                    res.append(temp)

    if a==1 :
        for yy in res :
            if len(yy)==cnt:
                realres.append(yy)
        return realres
    else :
        cnt+=1
        return makekmer(a-1)
```
