```python
f=open("C:\\Users\\cofls\\Downloads\\rosalind_grph.txt","r")

datas=[]
names=[]

tempdata=''
for i in f.readlines() :
    if ">" not in i :
        tempdata+=i.strip()
    else :
        datas.append(tempdata)
        tempdata=''
        names.append(i.strip().strip(">"))
datas.append(tempdata)
tempdata=''
del datas[0]

f.close()
```

```python
k=3
res=[]

for a in range(len(names)) :
    for b in range(a+1,len(names)) :
        #print(datas[a][-k:], datas[b][:k] )
        if datas[a][-k:]==datas[b][:k] :
            res.append(str(names[a])+" "+names[b])
        if datas[a][:k]==datas[b][-k:] :
            res.append(str(names[b]+" "+names[a]))
```

```python
fout=open("240204_result.txt","w")

for r in res :
    fout.write(r+"\n")
    
fout.close()
```
