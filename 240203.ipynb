```python
f=open("C:\\Users\\cofls\\Downloads\\rosalind_cons.txt","r")

datas=[]
tempdata=''
for i in f.readlines() :
    if ">" not in i :
        tempdata+=i.strip()
    else :
        datas.append(tempdata)
        tempdata=''
datas.append(tempdata)
tempdata=''
        
f.close()
del datas[0]
```

```python
datas
```

```python
res=[]
templine=[]
temp=0
ntlist=["A","C","G","T"]

for nt in ntlist :
    for x in range(len(datas[0])) :
        for y in range(len(datas)) :
            if datas[y][x]==nt :
                temp+=1
        templine.append(temp)
        temp=0
    res.append(templine)
    templine=[]
    
con=''
for j in range(len(datas[0])) :
    tempcompare=[res[0][j],res[1][j],res[2][j],res[3][j]]
    
    idx=tempcompare.index(max(tempcompare))
    con+=ntlist[idx]
    
```

```python
con
```

```python
printing=con+"\n"

forsave=''

for n in range(4) :
    printing+=ntlist[n]+": "
    for nn in range(len(datas[0])) :
        forsave+=str(res[n][nn])+" "
    printing+=forsave+"\n"
    forsave=''
```

```python
ff=open("240203_result.txt","w")
ff.write(printing)
ff.close()
```
