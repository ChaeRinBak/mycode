```python
data=datas[0]
data=data.replace("T","U")

o1=data
o2=data[1:]
o3=data[2:]
back=data[::-1]
back=back.replace("A","aa").replace("C","cc").replace("G","gg").replace("U","uu")
back=back.replace("aa","U").replace("gg","C").replace("cc","G").replace("uu","A")

o4=back    # backward frame은 complement함!!!!!!!!!!!!
o5=back[1:]
o6=back[2:]
orflist=[o1,o2,o3,o4,o5,o6]

tempres=''
res=[]
st=0
ts=[]
startidx=[]
endidx=[]

for y in orflist:
    for x in range(0,len(y),3) :
        if len(y[x:x+3])!=3 :
            #print(4,x,len(y[x:x+3]))
            st=0
            break
        if codon[y[x:x+3]]=="Stop" and st==1 :
            for p in range(len(ts)) :
                endidx.append(x)
            ts=[]
            st=0
            print(1,len(startidx),len(endidx))
        elif codon[y[x:x+3]]=="M":
            startidx.append(x)
            ts.append(x)
            st=1
    st=0
    ts=[]
        
    print(2,len(startidx), len(endidx))
    while len(startidx)!=len(endidx) :
        del startidx[-1]
    
    for z in range(len(startidx)) :
        for zz in range(startidx[z],endidx[z],3) :
            tempres+=codon[y[zz:zz+3]]
        res.append(tempres)
        tempres=''
   
    startidx=[]
    endidx=[]
    #print('---')
    
```
