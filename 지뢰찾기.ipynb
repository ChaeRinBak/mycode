```python
# 211129

board=[]
xin=[]
yin=[]
o=0
bb=[]

def bnc() : 
    bomnum=0
    if yin[o]==0 and xin[o]==0 :
        if board[(-yin[o])][xin[o]+1] :
            bomnum+=1
        if board[(-yin[o])+1][xin[o]] :
            bomnum+=1
        if board[(-yin[o])+1][xin[o]+1] :
            bomnum+=1
    if yin[o]==0 and xin[o]==4 :
        if board[(-yin[o])][xin[o]-1] :
            bomnum+=1
        if board[(-yin[o])+1][xin[o]] :
            bomnum+=1
        if board[(-yin[o])+1][xin[o]-1] :
            bomnum+=1
    if yin[o]==0 and 0<xin[o]<4 :
        if board[(-yin[o])][xin[o]+1] :
            bomnum+=1
        if board[(-yin[o])][xin[o]-1] :
            bomnum+=1
        if board[(-yin[o])+1][xin[o]] :
            bomnum+=1
        if board[(-yin[o])+1][xin[o]-1] :
            bomnum+=1
        if board[(-yin[o])+1][xin[o]+1] :
            bomnum+=1
    if yin[o]==-4 and 0<xin[o]<4 :
        if board[(-yin[o])][xin[o]+1] :
            bomnum+=1
        if board[(-yin[o])][xin[o]-1] :
            bomnum+=1
        if board[(-yin[o])-1][xin[o]] :
            bomnum+=1
        if board[(-yin[o])-1][xin[o]-1] :
            bomnum+=1
        if board[(-yin[o])-1][xin[o]+1] :
            bomnum+=1
    if xin[o]==0 and 0<-yin[o]<4 :
        if board[(-yin[o])+1][xin[o]] :
            bomnum+=1
        if board[(-yin[o])-1][xin[o]] :
            bomnum+=1
        if board[(-yin[o])-1][xin[o]+1] :
            bomnum+=1
        if board[(-yin[o])][xin[o]+1] :
            bomnum+=1
        if board[(-yin[o])+1][xin[o]+1] :
            bomnum+=1
    if xin[o]==4 and 0<-yin[o]<4 :
        if board[(-yin[o])+1][xin[o]] :
            bomnum+=1
        if board[(-yin[o])-1][xin[o]] :
            bomnum+=1
        if board[(-yin[o])-1][xin[o]-1] :
            bomnum+=1
        if board[(-yin[o])][xin[o]-1] :
            bomnum+=1
        if board[(-yin[o])+1][xin[o]-1] :
            bomnum+=1        
    if yin[o]==-4 and xin[o]==0 :
        if board[(-yin[o])][xin[o]+1] :
            bomnum+=1
        if board[(-yin[o])-1][xin[o]] :
            bomnum+=1
        if board[(-yin[o])-1][xin[o]+1] :
            bomnum+=1
    if yin[o]==-4 and xin[o]==4 :
        if board[(-yin[o])][xin[o]-1] :
            bomnum+=1
        if board[(-yin[o])-1][xin[o]] :
            bomnum+=1
        if board[(-yin[o])-1][xin[o]-1] :
            bomnum+=1
    if (0>yin[o]>-4) and (0<xin[o]<4) :
        if board[(-yin[o])-1][xin[o]-1] :
            bomnum+=1
        if board[(-yin[o])-1][xin[o]] :
            bomnum+=1
        if board[(-yin[o])-1][xin[o]+1] :
            bomnum+=1
        if board[(-yin[o])][xin[o]+1] :
            bomnum+=1
        if board[(-yin[o])][xin[o]-1] :
            bomnum+=1
        if board[(-yin[o])+1][xin[o]+1] :
            bomnum+=1
        if board[(-yin[o])+1][xin[o]] :
            bomnum+=1
        if board[(-yin[o])+1][xin[o]-1] :
            bomnum+=1
    bb.append(bomnum)
    return bb

import random
size=5
board=[]
tt=0
t=0
xin=[]
yin=[]

for x in range(size) :
    a=[]
    for y in range(size) :
        a.append(False)
    board.append(a)

for r in range(size) :
    for c in range(size) :  # c가 가로, r이 세로
        if (random.random()<0.3) :
            board[r][c] = True
            tt+=1
# print(board)

for r in range(size) :
    for c in range(size) :
        print('.', end='')
    print()
    
for o in range(size**2 +1) :
    if 25-tt==t :
        print('You win!')
        for r in range(size) :
            for c in range(size) :
                if board[r][c] :
                    print('#', end='')  # true가 #이고 폭탄
                else :
                    print('.', end='')
            print()
        break
    print('왼쪽 상단 (0,0)부터 우측 하단 (4,-4) 중 선택,  y<=0 주의 / 폭탄 수 : %s' %tt)
    xin.append(int(input('x 좌표 선택 : '))) 
    yin.append(int(input('y 좌표 선택 : ')))
    bnc()
    xy=list(zip(xin,yin))
    if board[-yin[o]][xin[o]] :  # 폭탄이면 사망
        print("Bomb!  You're dead!")  # true가 #이고 폭탄
        for r in range(size) :
            for c in range(size) :
                if board[r][c] :
                    print('#', end='')  # true가 #이고 폭탄
                else :
                    print('.', end='')
            print()
        break
    else : # 폭탄이 아닐 때 점이 폭탄수
        for r in range(size) :
            for c in range(size) :
                if (c,-r) in xy :
                    print(bb[xy.index((c,-r))], end='')  # true가 #이고 폭탄
                else :
                    print('.', end='')
            print()
    print()
    t+=1
```
