```python
# 230831

seq=input('뉴클레오타이드 시퀀스 : ')
ew=int(input('palindromic seq의 최소 nt 수'))

a=2 # 중앙 nt 개수
c=10 # 움직이는 중심점
result='' # palindormic한 거 표시
cnt=0 # index 표시

for a in range(2,len(seq)) :
    for c in range(len(seq)) :
        if c+a-1>=len(seq) :
            break

        central=seq[c:c+a]

        compareseq1=seq[:c]
        compareseq2=seq[c+a:]

        compareseq1=compareseq1[::-1]

        #print(compareseq1)
        #print(compareseq2)


        for i in range(min(len(compareseq1),len(compareseq2))) :
            if compareseq1[i]==compareseq2[i] :
                result+=compareseq1[i]
                if (i== (min(len(compareseq1),len(compareseq2))-1)) and (len(result)>=ew) :
                    #print('central=',central, c+1) # palindromic한 기준이 되는 중앙 seq
                    print('central', c+1)
                    print(':',result, c-cnt+1)
                    result=''
            elif len(result)>=ew :
                print('central', c+1)
                #print('central=',central, c+1)
                print(':',result, c-cnt+1)
                result=''
            else :
                result=''
            cnt+=1
        result=''
        cnt=0
```
